```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'sales' is your DataFrame
sales = pd.read_excel("/content/drive/MyDrive/Superstore Marketing Data.xlsx")

# Grouping by 'Education' and calculating the average 'Income'
education_income = sales.groupby('Education')['Income'].mean().reset_index()

# Plotting the line graph
plt.figure(figsize=(10, 6))
plt.plot(education_income['Education'], education_income['Income'], marker='o')
plt.title('Average Income by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Average Income')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
sales = pd.read_excel("/content/drive/MyDrive/Superstore Marketing Data.xlsx")

# Grouping data by Marital status and counting complaints
complaints_by_marital = sales.groupby('Marital_Status').agg({'Complain': 'sum'}).reset_index()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(complaints_by_marital['Marital_Status'], complaints_by_marital['Complain'], marker='o')
plt.title('Complaints by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Number of Complaints')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

## Task 3 - Probability Distributions üé≤

- **Identify variables that could follow specific probability distributions (e.g., Binomial, Normal). Calculate probabilities and expected values based on these distributions.**



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
from scipy import stats

# Calculate mean and standard deviation of income
mean_income = sales['Income'].mean()
standard_deviation = sales['Income'].std()

# Define the upper limit for probability calculation
# For example, let's calculate the probability of income being less than or equal to 75000
upper_limit = 75000

# Calculate the probability using the cumulative distribution function (cdf)
probability = stats.norm.cdf(upper_limit, mean_income, standard_deviation)

# Print the probability
print(f"The probability of income being less than or equal to {upper_limit} is: {probability}")
```

```python
# Creating a normal distribution plot in Python is straightforward using libraries like Matplotlib and NumPy. Here's a simple example to get you started:

import numpy as np
import matplotlib.pyplot as plt

# Generate data for a normal distribution
mean = 0
std_dev = 1
data = np.random.normal(mean, std_dev, 1000)

# Create the histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Plot the normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = np.exp(-0.5*((x - mean)/std_dev)**2) / (std_dev * np.sqrt(2 * np.pi))
plt.plot(x, p, 'k', linewidth=2)

plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()


# This code will generate a histogram of the data and overlay a normal distribution curve. Feel free to adjust the mean, std_dev, and other parameters to fit your specific needs. Enjoy plotting!
```
